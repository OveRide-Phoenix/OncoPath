## Import Library

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torch.utils.tensorboard import SummaryWriter
import os
import glob
import shutil
import cv2
from sklearn.model_selection import train_test_split
from torchvision.transforms import ToTensor, Compose, Normalize, RandomAffine, ColorJitter, Resize
from torchvision.models import efficientnet_b6, EfficientNet_B6_Weights
from sklearn.metrics import accuracy_score, recall_score, f1_score
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
```

## Class Dataset

```python
class BreastCancerDataset(Dataset):
    def __init__(self, root='/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT', is_train=True, transform=None):
        self.categories = ["benign", "malignant", "normal"]
        self.transform = transform
        
        paths = []
        labels = []
        
        for i, category in enumerate(self.categories):
            masks = glob.glob("{}/*_mask.png".format(os.path.join(root, category)))
            paths.extend(masks)
            labels.extend([i] * len(masks))
        
        train_paths, test_paths, train_labels, test_labels = train_test_split(paths, labels, test_size=0.15, random_state=42)
        
        if is_train:
            self.paths, self.labels = train_paths, train_labels
        else:
            self.paths, self.labels = test_paths, test_labels
            
    def __len__(self):
        return len(self.labels)
    
    def __getitem__(self, index):
        mask = cv2.imread(self.paths[index])
        mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)
        
        label = self.labels[index]
        
        if self.transform:
            mask = self.transform(mask)
            
        return mask, label
```

## Augumentation

```python
size = 224
train_transform = Compose([
    ToTensor(),
    RandomAffine(
        degrees=(-5, 5),
        translate=(0.15, 0.15),
        scale=(0.85, 1.15),
        shear=10
    ),
    ColorJitter(
        brightness=0.125,
        contrast=0.5,
        saturation=0.5,
        hue=0.05
    ),
    Resize((size, size)),
    Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

test_transform = Compose([
    ToTensor(),
    Resize((size, size)),
    Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

## UnNormalize

```python
class UnNormalize(object):
    def __init__(self, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]):
        self.mean = mean
        self.std = std
    
    def __call__(self, tensor):
        for t, m, s in zip(tensor, self.mean, self.std):
            t.mul_(s).add_(m)
        
        return tensor
unorm = UnNormalize()
```

```python
train_dataset = BreastCancerDataset(transform=train_transform)
test_dataset = BreastCancerDataset(is_train=False, transform=test_transform)

indexes = [4, 14, 24, 34, 44, 54, 64, 74, 84, 94]

fig = plt.figure(figsize=(20, 10))

for i, idx in zip(range(1, 11), indexes):
    fig.add_subplot(2, 5, i)
    mask, label = train_dataset[idx]
    plt.imshow(unorm(mask).permute(1, 2, 0))
    plt.title(f"Label: {train_dataset.categories[label]}")

plt.show()
```

```python
class AverageMeter(object):
    def __init__(self):
        self.reset()
    
    def reset(self):
        self.sum = 0
        self.count = 0
        self.avg = 0
    
    def update(self, val, n = 1):
        self.sum += val * n
        self.count += n
        self.avg = self.sum / self.count
```

## Show Loss

```python
def plot_segmentation_metrics(train_loss, val_loss, train_mean_iou=None, val_mean_iou=None, 
                              train_dice_score=None, val_dice_score=None):
    """
    Plot training and validation metrics (loss, mean IoU, Dice score) over epochs.

    Args:
    - train_loss (list): List of training loss values over epochs.
    - val_loss (list): List of validation loss values over epochs.
    - train_mean_iou (list, optional): List of training mean IoU values over epochs.
    - val_mean_iou (list, optional): List of validation mean IoU values over epochs.
    - train_dice_score (list, optional): List of training Dice score values over epochs.
    - val_dice_score (list, optional): List of validation Dice score values over epochs.
    """
    epochs = range(1, len(train_loss) + 1)
    
    plt.figure(figsize=(15, 10))
    
    # Plotting the Loss
    plt.subplot(2, 2, 1)
    plt.plot(epochs, train_loss, 'r', label='Training Loss')
    plt.plot(epochs, val_loss, 'b', label='Test Loss')
    plt.title('Training and Test Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    # Plotting the Mean IoU (if available)
    if train_mean_iou and val_mean_iou:
        plt.subplot(2, 2, 2)
        plt.plot(epochs, train_mean_iou, 'r', label='Training Mean IoU')
        plt.plot(epochs, val_mean_iou, 'b', label='Test Mean IoU')
        plt.title('Training and Test Mean IoU')
        plt.xlabel('Epochs')
        plt.ylabel('Mean IoU')
        plt.legend()
    
    # Plotting the Dice Score (if available)
    if train_dice_score and val_dice_score:
        plt.subplot(2, 2, 3)
        plt.plot(epochs, train_dice_score, 'r', label='Training Dice Score')
        plt.plot(epochs, val_dice_score, 'b', label='Test Dice Score')
        plt.title('Training and Test Dice Score')
        plt.xlabel('Epochs')
        plt.ylabel('Dice Score')
        plt.legend()
    
    plt.tight_layout()
    plt.show()
```

## Train

```python
def train(num_class, batch_size, lr, epochs, checkpoint, saved_model):
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    
    n_workers = os.cpu_count()
    train_dataset = BreastCancerDataset(transform=train_transform)
    train_loader = DataLoader(
        dataset=train_dataset,
        batch_size=batch_size,
        shuffle=True,
        num_workers = n_workers
    )
    
    test_dataset = BreastCancerDataset(is_train=False, transform=test_transform)
    test_loader = DataLoader(
        dataset=test_dataset,
        batch_size=batch_size,
        num_workers=n_workers
    )
    
    model = efficientnet_b6(weights=EfficientNet_B6_Weights)
    in_features = model.classifier[1].in_features
    model.classifier[1] = nn.Linear(in_features=in_features, out_features=num_class)
    model.to(device)
    
    optimizer = torch.optim.AdamW(params=model.parameters(), lr=lr, weight_decay=1e-4)
    criterion = nn.CrossEntropyLoss()
    
    num_iter = len(train_loader)
    scheduler = torch.optim.lr_scheduler.PolynomialLR(optimizer, epochs*num_iter, power=0.9)
    
    if os.path.isdir(checkpoint):
        shutil.rmtree(checkpoint)
    
    os.makedirs(checkpoint)
    
    if not os.path.isdir(saved_model):
        os.makedirs(saved_model)
    
    writer = SummaryWriter(checkpoint)
    
    train_loss = []
    test_loss = []
    train_loss_meter = AverageMeter()
    test_loss_meter = AverageMeter()
    best_acc = -1
    no_improvement_counter = 0
    patience = 10
    
    for epoch in range(epochs):
        model.train()
        progress_bar = tqdm(train_loader, colour='cyan')
        train_loss_meter.reset()
        test_loss_meter.reset()
        
        for i, (mask, label) in enumerate(progress_bar):
            mask, label = mask.to(device), label.to(device)
            
            output = model(mask)
            
            loss = criterion(output, label)
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            train_loss_meter.update(loss.item())
            
            progress_bar.set_description("Train | Epoch: {}/{} | Loss {:0.4f} | lr {}".format(epoch + 1, epochs, train_loss_meter.avg, optimizer.param_groups[0]['lr']))
            
            writer.add_scalar("Train/Loss", loss.item(), epoch * num_iter + i)
            writer.add_scalar("Train/Loss AVG", train_loss_meter.avg, epoch * num_iter + i)
            
            scheduler.step()
        
        model.eval()
        with torch.no_grad():
            progress_bar = tqdm(test_loader, colour='yellow')
            all_labels = []
            all_predictions = []
            
            for mask, label in progress_bar:
                mask, label = mask.to(device), label.to(device)
                
                output = model(mask)
                predictions = torch.argmax(output, dim=1)
                loss = criterion(output, label)
                
                test_loss_meter.update(loss.item())
                
                all_predictions.extend(predictions.tolist())
                all_labels.extend(label.tolist())
                
                progress_bar.set_description("Val | Loss {:0.4f}".format(train_loss_meter.avg))
                
            acc = accuracy_score(all_labels, all_predictions)
            recall = recall_score(all_labels, all_predictions,  average='macro')
            f1 = f1_score(all_labels, all_predictions,  average='macro')
            print(f"Val | Loss {test_loss_meter.avg:.4f} | Acc: {acc} | Recall: {recall} | F1-score: {f1}")
            
            writer.add_scalar("Val/Loss", test_loss_meter.avg, epoch)
            writer.add_scalar("Val/Acc", acc, epoch)
            writer.add_scalar("Val/Recall", recall, epoch)
            writer.add_scalar("Val/F1-score", f1, epoch)
            
            checkpoint = {
                "model_state_dict": model.state_dict(),
                "epoch": epoch + 1,
                "optimizer_state_dict": optimizer.state_dict(),
                "acc": acc
            }
            
            torch.save(checkpoint, os.path.join(saved_model, "last.pt"))
            
            if acc > best_acc:
                torch.save(checkpoint, os.path.join(saved_model, "best.pt"))
                best_acc = acc
                no_improvement_counter = 0
            else:
                no_improvement_counter += 1
            
            if acc == 1.0 and recall == 1.0 and f1 == 1.0:
                break
            
            train_loss.append(train_loss_meter.avg)
            test_loss.append(test_loss_meter.avg)
            
            if no_improvement_counter > patience:
                print("Early stopping!")
                break
    
    plot_segmentation_metrics(train_loss, test_loss)
```

```python
torch.cuda.empty_cache()
```

```python
num_class = 3
batch_size = 32
lr = 1e-3
epochs = 50
checkpoint = '/kaggle/working/tensorboard'
saved_model = '/kaggle/working/saved_model'
train(num_class, batch_size, lr, epochs, checkpoint, saved_model)
```

## Inference

```python
def pred_show_image_grid(model_pth, device, test_dataset, num_classes, idx, num_imgs):
    # Load model checkpoint
    checkpoint = torch.load(model_pth)
    
    # Initialize and load model state
    model = efficientnet_b6(weights=EfficientNet_B6_Weights)
    in_features = model.classifier[1].in_features
    model.classifier[1] = nn.Linear(in_features=in_features, out_features=num_classes)  # Corrected to num_classes
    model.to(device)
    
    model.load_state_dict(checkpoint['model_state_dict'])
    model.eval()
    
    # Initialize lists to store original masks, ground truths, and predictions
    orig_masks, ground_truths, pred_labels = [], [], []
    softmax = torch.nn.Softmax(dim=1)
    
    with torch.no_grad():
        for i in idx:
            ori_mask, orig_label = test_dataset[i]
            mask = ori_mask.unsqueeze(0)  # Add batch dimension

            # Predict mask
            output = model(mask.to(device))
            prob = softmax(output)
            predicted_prob, predicted_class = torch.max(prob, dim=1)
            score = predicted_prob[0].item() * 100  # Extract the probability value

            # Append the original mask, ground truth, and predicted label
            orig_masks.append(ori_mask.permute(1, 2, 0))
            ground_truths.append(test_dataset.categories[orig_label])
            pred_labels.append(test_dataset.categories[predicted_class.item()] + f" {score:0.2f}%")
    
    # Plot images and masks
    fig = plt.figure(figsize=(22, 11))

    for i, (mask, label, pred) in enumerate(zip(orig_masks, ground_truths, pred_labels)):
        ax = fig.add_subplot(2, 5, i + 1)  # Indexing starts at 1, so we add 1
        ax.imshow(mask)
        ax.set_title(f"Label: {label}", fontsize=15, loc='left')
        ax.set_xlabel(f"Pred: {pred}", fontsize=15, loc='right')

    plt.show()
```

```python
#Note: You must train the model before using this param
model_pth = "/kaggle/working/saved_model/best.pt"
num_classes = 3
device = "cuda" if torch.cuda.is_available() else "cpu" 
print("\n")
indexs = [4, 14, 24, 34, 44, 54, 64, 74, 84, 94]
print("indexs =", indexs)
print("\n")

#Show images
pred_show_image_grid(model_pth, device, test_dataset, num_classes, indexs, num_imgs= 10)
```
